```python
#from plotnine import ggplot,geom_col,aes,facet_grid,labs
```

```python
graph=(
ggplot(df)
+aes(x="Chromosome",y="Count_in_millions")
+labs(x="Chromosome",y="Cytosine count in millions", title= "Methylatable C distribution in A. thaliana")
)
print(graph)
```

```python
graph1=graph+geom_col()
print(graph1)
```

```python
graph2=(graph1+geom_col()+aes(fill="Context"))
print(graph2)
```

```python
graph3=(graph+geom_col(position="dodge")+aes(fill="Context"))
print(graph3)
```

```python
graph4=(graph3+facet_grid("Context ~ ."))
print(graph4)
```

```python
graph5=(graph3+facet_grid(".~Context"))
print(graph5)
```

```python
# from plotnine import scale_fill_manual
graph6=graph5+scale_fill_manual(values=["orange","purple","darkgreen"])
print(graph6)
```

```python
from plotnine import theme_bw
graph7=graph6+theme_bw(base_size=9)
print(graph7)
```
