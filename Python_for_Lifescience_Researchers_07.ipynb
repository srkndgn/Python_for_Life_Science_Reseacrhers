```python
# Reverse complement
# DNA has 2 strands
# made of 4 nucleotides = A, T, C, G
# On opposite strand - A pairs with T and G pairs with C
# 2 strands are antiparallel
# 5' ATGCATCG 3'
# 3' TACGTAGC 5'
```

```python
dna="ATGCATGC"
```

```python
# Replacing a character/substring with another character/substring
dna=dna.replace("A","T")
print(dna)
dna=dna.replace("T","A")
print(dna)
```

```python
dna="ATGCATGC"
```

```python
print(dna)
dna=dna.replace("A","t")
dna=dna.replace("T","a")
dna=dna.replace("G","c")
dna=dna.replace("C","g")

complement=dna.upper()
print(complement)
reverse_complement=complement[::-1]
print(reverse_complement)
```

```python
# Generating the complement of a sequence

dna=input("Submit a DNA sequence").upper()
print(dna)

complement=dna.replace("A","t").replace("T","a").replace("G","c").replace("C","g")
print(complement)
complement=complement.upper()
print(complement)
```

```python
# Reverse of the complement
reverse_complement=complement[::-1]
print(reverse_complement)
```
