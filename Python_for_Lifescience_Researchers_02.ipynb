```python
#Tuples are very similar to list - except
#1. They are declared with round brackets
#2. They are not mutable or they can not be changed
```

```python
#Declaring a tuple
cd_list=["DNA","RNA","Protein"] # list - square brackets
cd_tuple=("DNA","RNA","Protein") # tuple - round brackets
print(type(cd_list))
print(type(cd_tuple))
```

```python
#Tuples are similar to list - zero indexing except that ....
print(cd_tuple)
print(cd_tuple[1])
print(cd_tuple[1:])
print(cd_tuple[:1])

```

```python
#Lists can be modified - mutable
print(cd_list)
cd_list[0]="Nanopore"
print(cd_list)
```

```python
#Tuples can not be modified - not mutable
print(cd_tuple)
cd_tuple[0]="Nanopore"
print(cd_tuple)
```

```python
#Like lists you can iterate over a tuple
squencers=("Illumina","IonTorrent","Nanopore","PacBio")
for i in squencers:
  print(i)
```

```python
#Looking for a composition of a protein sequence
#List of amino acids as tuple
amino_acids=("A", "R", "N", "D", "C", "Q", "E", "G", "H", "I", "L", "K", "M", "F", "P", "S", "T", "W", "Y", "V")

#take the protein sequence as input
protein=input("Submit a protein sequence ").upper()
print(protein)

#Use for loop to access individual amino acids and query the sequence
for i in amino_acids:
  count1=protein.count(i)
  percent=(count1/len(protein))*100
  percent2=round(percent,2)
  print(i,percent2)
```
