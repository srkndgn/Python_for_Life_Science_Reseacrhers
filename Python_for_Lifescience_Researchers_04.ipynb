```python
#Importing library
import pandas as pd
```

```python
#Open a fasta file; read the sequence and determine its composition and the GC content , and write it to a new file
#Read Write Append
#Opening file for reading (file1) and writing(file2)

read1=open("gapdh.txt","r")    #read mode
write1=open("Results.txt","a") #append mode
```

```python
#initializing flags
count=0
seq=""
```

```python
#Reading file line by line, removing the new line character and joining
#all lines exept the def line

for line in read1:
  line=line.rstrip("\n") #removing new line chracter

  count +=1
  if count == 1:
    write1.write(line) #writeing the def line

  else:
    seq=seq+line #concatenating sequence line
    print(seq,"\n\n")
```

```python
length1=len(seq)
length1
write1.write("\n" + "length of the sequence is: "+str(length1) + "\n")
```

```python
#Calculating the composition of the DNA Sequence

nucleotide=("A","T","C","G","N")
gc=0

for x in nucleotide:
  n_count=seq.count(x)
  write1.write(str(x)+"\t"+str(n_count)+"\n")
  if x=="G" or x=="C":
    gc=gc+n_count
  print(x,n_count)

percent_gc=(gc/length1)*100
write1.write("\n" + "GC content=" + str(percent_gc) +"\n")
```

```python
read1.close() # close the file and check inside
write1.close()
```
