```python
#Write a code to calculate total number of DNA molecules if you start with x molecules of DNA and do n cycles of PCR
#Take x and n as input
# Input values
x = int(input("Enter the initial number of DNA molecules: "))
n = int(input("Enter the number of PCR cycles: "))

# Calculate the total number of DNA molecules
total_molecules = x * (2 ** n)

# Display the result
print("Total number of DNA molecules after", n, "cycles of PCR:", total_molecules)
```

```python
# Input values
x = int(input("Enter the initial number of DNA molecules: "))
n = int(input("Enter the number of PCR cycles: "))

# Initialize a variable to keep track of the current number of DNA molecules
current_molecules = x

# Perform PCR cycles using a for loop
for cycle in range(1, n + 1):
    current_molecules *= 2
    print("After cycle", cycle, ":", current_molecules, "DNA molecules")

# Display the final total
print("Total number of DNA molecules after", n, "cycles of PCR:", current_molecules)
```

```python
#Calculate number of purines and pyrymidines

dna=input("Submit a DNA sequence ")
dna=dna.upper()
print(dna)
```

```python
#initializing counters
purine=0; pyrimidine=0;

#setting up iteration - for loop
for i in dna:
  if i=="A" or i=="G":
    purine=purine+1

  else:
    pyrimidine=pyrimidine+1

print("Total Purines ", purine, "Total pyrimidines", pyrimidine)
```

```python
#Calculate number of purines and pyrymidines

dna=input("Submit a DNA sequence ")
dna=dna.upper()
print(dna)


#initializing counters
purine=0; pyrimidine=0; un=0

#setting up iteration - for loop
for i in dna:
  if i=="A" or i=="G":
    print(i,"Purine")
    purine+=1
    #purine=prine+1

  elif i=="C" or i=="T":
    print(i,"Pyrimidine")
    pyrimidine=pyrimidine+1

  else:
    print(i,"Unknown")
    un+=1

print("Total Purines = ",purine, "Total pyrimidines = ",pyrimidine, "Unknown = ",un)
```

```python
# Sum of digits in a number submitted by the user

# Input a number from the user
num_str = input("Enter a number: ")

# Initialize a variable to store the sum of digits
sum_of_digits = 0

# Iterate through each character in the input string
for digit_char in num_str:
    # Check if the character is a digit
    if digit_char.isdigit():
        # Convert the character to an integer and add it to the sum
        sum_of_digits += int(digit_char)

# Display the sum of digits
print("Sum of digits in the entered number:", sum_of_digits)

```
