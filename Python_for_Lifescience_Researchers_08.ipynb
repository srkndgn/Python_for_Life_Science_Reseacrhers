```python
# Reading fasta file and calculating its composition for each file

import glob

filenames=glob.glob("*.fasta")
print(filenames)
```

```python
nucleotide=("A","T","C","G","N")
file2=open("Results.txt", "a")
```

```python
#The first loop iterates filenames
for i in filenames:
#intializing flags
    count=0
    seq=""
    gc=0
    percent_gc=0
    file1=open(i,"r")

#The inner loop
#Reading file line by line, removing the new line
#character and joining all lines except the def line

    for line in file1:
        line = line.rstrip("\n") #removing new line character
        count +=1
        if count==1:
            file2.write(line +"\n") #writing the def line
        else:
            seq=seq+line #concatenating sequence lines

#Calculating composition and gc percent
    length1=len(seq)
    for x in nucleotide:
        n_count=seq.count(x)
        file2.write(str(x)+str(n_count)+"\t")
        if x=="G" or x=="C":
            gc=gc+n_count

    percent_gc=(gc/length1)*100
    file2.write("\n" +"GC content="+ str(percent_gc)+"\n")
    file1.close()

file2.close()
```
